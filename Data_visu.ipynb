```python
#4How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt


data = {
    'Math': [85, 78, 90, 88, 76],
    'Science': [92, 80, 89, 94, 78]

}


df = pd.DataFrame(data)


corr_matrix = df.corr()


sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
#10Create a histogram using Seaborn to visualize a distribution
import seaborn as sns
Mark={'Name':['ting','Kanha','Deva','Madhab','Adi'],'mark':[90,60,25,70,66]}
df=pd.DataFrame(Mark)
sns.histplot(df['mark'],bins=5,color='yellow')
plt.title("Mark Distrubution")
plt.xlabel('Mark')
plt.ylabel('Frequency')
plt.show()
```
